```python
import sacrebleu

# Đã qua bước tách từ (word segmentation)
hyp = "Con mèo_máy ngồi trên cái chiếu"
ref = "Con mèo_máy ngồi trên cái chiếu"

ter_metric = sacrebleu.metrics.TER()

# SacreBLEU yêu cầu ref là list của list
score = ter_metric.corpus_score([hyp], [[ref]])

print(f"Bản dịch máy: {hyp}")
print(f"Bản dịch chuẩn: {ref}")
print(f"TER Score: {score.score:.2f}")
```
